```python
import numpy as np
import UPAFuzzySystems as UPAfs
```

```python
OilPrice = UPAfs.fuzzy_universe('Oil Price',np.linspace(5,30,50),'continuous')
OilPrice.add_fuzzyset('Low','trapmf',[5,5,10,14])
OilPrice.add_fuzzyset('Good','trimf',[10,15,22])
OilPrice.add_fuzzyset('High','trapmf',[16,22,30,30])
```

```python
OilPrice.view_fuzzy()
```
